```python
import pandas as pd  

# Carregar os dados principais  
print(f"O arquivo carregado deve ser obrigatoriamente .csv,\ncom colunas padrão definidas.")
print("Utilize apenas o modelo especificado!")

should_exit = False

try:
    arq = input("Escreva o nome do arquivo(Ex: 'arquivo.csv'): ")
    print(arq)
    df = pd.read_csv(arq, sep=",")
    print("Sua tabela: ")
    display(df) 
except:
    ValueError or TypeError
    print("Utilize um arquivo valido ou modifique o nome.")
    should_exit = True
 

# Extraindo valores das colunas  
name = df["Nome"].values  
cod = df["Identificador"].values  
x = df["xM"].values  
y = df["yM"].values  
z = df["espessura (mm)"].values  
quant = df["quantidade"].values  

# Inicializar uma lista para armazenar os resultados  
result = []  

# Loop para iterar sobre os dados  
for i in range(len(df)):  
    f = "steelsheets.csv"  
    steelsheets = pd.read_csv(f, sep=",")   
    
    for _, row in steelsheets.iterrows():  
        espessura = row["ESPESSURAMM"]  
        kg = row["PESOKGM2"]  
        
        # Verificar se a espessura corresponde  
        if espessura == z[i]:  
            # Calcular peso  
            peso = x[i] * y[i] * kg * quant[i]  # Cálculo do peso  
            # Adicionar resultado à lista  
            result.append({  
                "Peça": name[i],  # Usar o índice i para pegar o nome correto  
                "Codigo": cod[i],  # Usar o índice i para pegar o código correto  
                "Espessura": espessura,  
                "Peso KG [total]": round(peso, 2),  # Arredondar para 2 casas decimais  
                "Dimensões": f"{x[i]} x {y[i]}",  
                "Quantidade": quant[i]  
            })  

# Criar um DataFrame com os resultados  
result_df = pd.DataFrame(result)  

# Reorganizar as colunas para exibição  
result_df = result_df[["Peça", "Codigo", "Espessura", "Peso KG [total]", "Dimensões", "Quantidade"]]  

# Exibir a tabela com os resultados 
print("Resultados: ")
display(result_df)
```
